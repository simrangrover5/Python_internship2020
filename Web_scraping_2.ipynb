```python
def download_image(link):
    name = link.split("/")[-1]
    p = requests.get(link)
    if p.status_code == 200:
        with open(f"{name}",'wb') as fp:
            fp.write(p.content)
    else:
        print("Link is not responding")
    
for i in soup.select("img")[:5]:
    h = i['src']
    if not "http" in i['src']:
        s = "https://www.scrapinghub.com" + h
        download_image(s)
```

```python
def get_soupobject(link):
    page = requests.get(link)
    if page.status_code == 200:
        soup = BeautifulSoup(page.content,'lxml')
        return soup
    else:
        return "Invalid url"


url2 = "https://www.scrapinghub.com/data-driven-product/"
data = get_soupobject(url2)
```
