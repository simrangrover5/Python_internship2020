### Entity Relationship Diagram

    Entity (table) --> It can be anything real or it can be any object (student,employee,school,college,teacher,animal etc...)
    Relationship --> It is relation between two entities 
        Example --> entity 1 --> student
                    entity 2 --> teacher
                    teacher teaches student
                    relationship --> teaches
                    
    Diagram --> The symbols of entity and relations
